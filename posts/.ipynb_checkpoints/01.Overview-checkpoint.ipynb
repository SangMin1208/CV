##  1. Course overview

### A. Goal

* Understand what is computer vision/deep learning algorithms
* Understand how to implement image/video processing algorithms in
Python using OpenCV.
* Understand how to implement/train/test convolutional neural networks
(CNNs) and Transformer in PyTorch.
