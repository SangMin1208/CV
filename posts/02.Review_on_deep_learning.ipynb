## 1. Deep vs. Machine learning

### A. Shallow learning vs. Deep learning

`-` Shallow learning

* only has one layer
* feature : vector

`-` Deep learning

* has more than one layer
* use Gradient Descent
* over fitting problem

`-` Extract optimal representation via End-to-End learning

## 2. CNNs(Convolutional Neural Networks)

`#`  RGB image $\rightarrow$ Any Differentiable Layers(learnability) $\rightarrow$ Semantic labels

### A. Differentiable layers and Gradient descent

`#` $I$ $\rightarrow$ **layer $h$** $\rightarrow$ $X$ 

$\rightarrow$ **Any Differentiable intermediate Layer $f$** whose parameter is $\theta$ 

$\rightarrow$ $y$ $\rightarrow$ **layer $g$**(loss function) $\rightarrow$ Loss $L$(distance btw ground truth & prediction)

* forward rule(순전파) : $y = f(h(I);\theta)\:\:$ for $\:\: g(f(h(I);\theta))=L$
* backward rule(역전파) : $\frac{dL}{dI} = \frac{dX}{dI} \times \frac{dy}{dX} \times \frac{dL}{dy}$
* parameter update rule : $\theta^{new} =  \theta - \epsilon(\text{learning rate})\frac{dy}{d\theta}\times\frac{dL}{dy}$
